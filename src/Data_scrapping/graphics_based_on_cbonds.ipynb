```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import cbonds list
cbonds_path = r'/Users/gabbyvinco/Desktop/HISTORICAL_BONDS.csv'
cbonds = pd.read_csv(cbonds_path, index_col=None, header=0)
```

```python
t_f_green = cbonds["Green bonds"].map(
                   {'yes':True ,'no':False})
```

```python
gb = sum(t_f_green)
ngb = len(t_f_green) - gb
```

```python
data = [gb, ngb]
```

```python
label = ["Green Bond", "Not Green Bond"]
```

```python
plt.pie(data,labels=label,autopct='%1.1f%%')
```
